```python
# Importando as bibliotecas
import pandas as pd
import seaborn as sns
```

```python
# Trazendo os dados da planilha excel
vendas = pd.read_excel('varejo.xlsx')
```

```python
vendas.head(10)
```

```python
vendas.info()
```

```python
# Filtra a quantidade de valores unicos por idcanalvenda
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
# Alteração dos valores APP para Aplicativo
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
# Filtrando Quantidade de valores únicos para cada bandeira
vendas.groupby("bandeira").idcompra.nunique()
```

```python
# Filtrando Quantidade de valores únicos para cada departamento
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
# Filtrando Quantidade de valores únicos para cada estado
vendas.groupby("estado").idcompra.nunique()
```

```python
# Fazendo o replace dos dados na categoria departamento -> substituindo espaços vazios entre palavras por _.
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
# Deleta todas as linhas que contenham linhas nulas
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
# Alterando todos os dados nulos de estados por MS
vendas["estado"] = vendas["estado"].fillna("MS")
```

```python
vendas[vendas["Preço"].isnull()]
```

```python
# Alterando todos os preços nulos pelo valor da média dos preços
media_preco = vendas["Preço"].mean()

vendas["Preço"] = vendas["Preço"].fillna(media_preco)
```

```python
mediana_preco = vendas["Preço"].median()

mediana_preco
```

```python
vendas.head()
```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")

vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto[["Preço","Preço_com_frete"]].describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' & estado == 'SP' ").describe()
```

```python
# QUANTIDADE DE COMPRAS POR DEPARTAMENTOS
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# FILTRANDO A MÉDIA DO PREÇO DE FRETE DE ACORDO COM O DEPARTAMENTO

round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
# FILTRANDO A MÉDIA DO PREÇO DE FRETE DE ACORDO COM O DEPARTAMENTO
vendas_correto.groupby("Nome_Departamento").Preço_com_frete.mean().sort_values(ascending=False)
```

```python
# Quantidade de Vendas por Data

vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
# Criando nova coluna de mês
vendas_correto["mes"] = vendas_correto["Data"].dt.month
```

```python
# Quantidade de Vendas por Mês

vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
vendas_correto.head()
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
# Alterando o tipo do dado
cliente = cliente.astype({"renda": "float"})
cliente.info()
```

```python
# Fazendo a junção da tabela vendas_correto com tabela clientes
vendas_cliente = vendas_correto.merge(cliente, how= "left", on = "cliente_Log")
vendas_cliente.head()
```

```python
# MÉDIA DE RENDA PARA CADA TIPO DE CANAL DE VENDA

round(vendas_cliente.groupby("idcanalvenda").renda.mean().sort_values(ascending=False).reset_index(),2)
```

```python
agg_canal_renda = vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index()
agg_canal_renda
```

```python
# MÉDIA DE IDADE DOS CLIENTES POR BANDEIRA

agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
# Visualização da Idade média por Bandeira

# Configurando o tamanho da figura
fig = plt.figure(figsize = (7,4))

# Eixo (x,y)
plt.bar(agg_idade_bandeira["bandeira"],agg_idade_bandeira["idade"],color = "b")

# Rótulo no eixo y
plt.ylabel("Média de Idade")

# Título
plt.title("Idade média por Bandeira",loc = "center", fontsize = 18, fontweight = "bold",color = "black");
```

```python
# Visualização da Média de Renda por Canal de Venda

fig = plt.figure(figsize = (7,4))

plt.bar(agg_canal_renda["idcanalvenda"],agg_canal_renda["renda"], color = "purple")

plt.ylabel("Renda Média")

plt.title("Renda média por Canal de Venda",loc = "center", fontsize = 18,fontweight = "bold", color = "black");
```

```python
vendas_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data
```

```python
# Quantidade de Vendas por Data
fig = plt.figure(figsize = (12,5))
plt.xlabel("Data da Venda")
plt.ylabel("Quantidade de Vendas")
plt.plot(vendas_por_data["Data"],vendas_por_data["idcompra"],color = "g")
plt.title("Quantidade de Vendas por Data",loc = "center", fontsize = 18, fontweight = "bold");
plt.show()
```

```python
vendas_departamento = vendas_correto.groupby("Nome_Departamento").Preço_com_frete.mean().sort_values(ascending=False).reset_index()
vendas_departamento
```

```python
# Preço médio por Departamento
fig = plt.figure(figsize = (24,8))

x = vendas_departamento["Nome_Departamento"]
y = vendas_departamento["Preço_com_frete"]

plt.xlabel("Departamento")
plt.ylabel("Preço com Frete")

# Altera a rotação dos rótulos
plt.xticks(rotation = 45)

plt.bar(x,y);
```
